## 1. Collaborative Filtering, Kernels, Linear Regression

```python
# 1.1 
lamb=1
U_0=np.array([6,0,3,6])
V_0=np.array([4,2,1])
X_0=np.outer(U_0,V_0)
X_0
```

```python
# 1.2 squared error term and regularization term in J(U,V)
Y = np.array([[     5, np.nan,      7],
              [np.nan,      2, np.nan],
              [     4, np.nan, np.nan],
              [np.nan,      3,      6]])

se = np.sum((np.nan_to_num(Y-X_0))**2)/2
reg= (lamb/2)*(np.sum(U_0**2)+np.sum(V_0**2))
print(f'squared error term={se}, regularization term={reg}')
```
